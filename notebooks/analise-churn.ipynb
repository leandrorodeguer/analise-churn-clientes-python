Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

 Passo 1: Importar a base de dados

```python
import pandas as pd

df_clientes = pd.read_csv("../data/cancelamentos.csv")
```

 Passo 2: Visualizar a base de dados
 colunas inúteis - remoção de informações que não me ajudam

```python
df_clientes = df_clientes.drop(columns="CustomerID")
display(df_clientes)
```

 Passo 3: Corrigir as falhas da base de dados

```python
display(df_clientes.info())
```

 Valores vazios
    
    excluir as linhas que têm valores vazios

```python
df_clientes = df_clientes.dropna()

display(df_clientes.info())
```

 Passo 4: Análise inicial dos cancelamentos

    Quantas pessoas cancelaram e quantas não cancelaram

```python
display(df_clientes["cancelou"].value_counts())
```

 Em percentual

```python
display(df_clientes["cancelou"].value_counts(normalize=True))
```

```python
display(df_clientes["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

 Passo 5: Análise das causas dos cancelamentos (como as colunas da base impactam no cancelamento)
   
     gráficos/dashboards

```python
import plotly.express as px
```

 Criar e exibir os gráficos

```python
import os
os.makedirs("imagens", exist_ok=True)

for coluna in df_clientes.columns:
    grafico = px.histogram(df_clientes, x=coluna, color="cancelou")
    grafico.show()
    grafico.write_image(f"imagens/{coluna}.png")
```

### Clientes do contrato mensal TODOS cancelam
     ofercer desconto nos planos anuais e trimestrais
### Clientes com mais de 4 ligações ao call center apresentam maior probabilidade de cancelamento
     criar um processo para resolver o problema do cliente em no máximo 3 ligações
### Clientes que atrasaram mais de 20 dias, cancelaram
     política de resolver atrasos em até 10 dias (equipe financeira)

# Simulação de aplicação de estratégias de retenção com base nos insights identificados

```python
df_clientes = df_clientes[df_clientes["duracao_contrato"]!="Monthly"]
df_clientes = df_clientes[df_clientes["ligacoes_callcenter"]<=4]
df_clientes = df_clientes[df_clientes["dias_atraso"]<=20]

display(df_clientes["cancelou"].value_counts())

```

 Em percentual

```python
display(df_clientes["cancelou"].value_counts(normalize=True))
```
